```python
from ultralytics import YOLO

# 1. Tải một model YOLOv8 (ví dụ: yolov8n.pt - bản 'nano' rất nhẹ)
model = YOLO('yolov8x.pt')

# 2. Chạy tracking
# - 'source' là video đầu vào
# - 'tracker' chỉ định file config của ByteTrack (có sẵn trong thư viện)
# - 'save=True' để lưu video kết quả
results = model.track(source='/kaggle/working/3d_simulator_20s.mp4', tracker='bytetrack.yaml', save=True, conf=0.4)
```
